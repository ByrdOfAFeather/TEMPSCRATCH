```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer

model = AutoModelForSequenceClassification.from_pretrained("/home/byrdofafeather/ByrdOfAFeather/TEMPSCRATCH/checkpoint-3000")
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
from transformers_interpret import SequenceClassificationExplainer

cls_explainer = SequenceClassificationExplainer(
    model, tokenizer
)

word_attributions = cls_explainer("I am a historian and engineer, here is my question: what year was the first engineer born?")
```

```python
cls_explainer.visualize("distilbert_negative_attr.html")
```
